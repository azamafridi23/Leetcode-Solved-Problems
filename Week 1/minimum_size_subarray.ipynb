Technique: Sliding Window
Link: https://www.youtube.com/watch?v=aYqYMIqZx5s



```python
def solution(arr,target):
    left, total = 0,0
    result = float("inf")

    for r in range(len(arr)):
        total += arr[r]
        while total>=target:
            result = min(r-left+1,result)
            total-=arr[left]
            left+=1    
    
    return 0 if result == float("inf") else result

arr= [2,3,1,2,4,3]
ans = solution(arr,7)
print(ans)
```
